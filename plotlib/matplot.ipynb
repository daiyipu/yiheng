```python
import  matplotlib.pyplot  as plt
import pandas as pd
import numpy as np
```

```python
# 绘制折线图
x = [1, 2, 3, 4, 5]
x1 = [1, 3, 2, 4, 5]
y = [5, 7, 2, 8, 4]

plt.rcParams['font.family'] = 'SimHei'
plt.plot(x, y, marker='o')
plt.plot(x1, y, marker='o')
plt.xlabel('X轴标签')
plt.ylabel('Y轴标签')
plt.title('折线图示例')
# plt.show()
```

```python
x1 = [1, 2, 3, 4, 5]
x2 = [2, 4, 5, 6, 7]
y = [5, 7, 2, 8, 4]

plt.scatter(x1, y, color='red', marker='o')
plt.scatter(x2, y, color='blue', marker='o')
plt.xlabel('X轴标签')
plt.ylabel('Y轴标签')
plt.title('散点图示例')
# plt.show()
```

```python
#  条形图
categories = ['A', 'B', 'C', 'D']
values1 = [20, 45, 30, 15]
values2 = [15, 23, 17, 10]

fig, ax = plt.subplots()
bars1 = ax.bar(categories, values1)
bars2 = ax.bar(categories, values2)

# 在每个条形图上添加文本
for bar in bars1:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), 
            ha='center', va='bottom')

# 在每个条形图上添加文本
for bar in bars2:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), 
            ha='center', va='bottom')

# plt.bar(categories, values1, color='blue')
# plt.bar(categories, values2, color='red')
plt.xlabel('类别')
plt.ylabel('值')
plt.title('条形图')
# plt.show()
```
