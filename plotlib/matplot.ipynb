#### 基本图表函数
- plot(): 绘制线图。
- scatter(): 绘制散点图。
- bar(): 绘制垂直条形图。
- barh(): 绘制水平条形图。
- hist(): 绘制直方图。
- boxplot(): 绘制箱型图。
- violinplot(): 绘制小提琴图。
- pie(): 绘制饼图。
- hexbin(): 绘制六边形箱图。

#### 坐标轴和图形控制
- xlabel(): 设置 x 轴标签。
- ylabel(): 设置 y 轴标签。
- title(): 设置图形标题。
- legend(): 添加图例。
- grid(): 显示或隐藏网格线。
- xlim(): 设置 x 轴的显示范围。
- ylim(): 设置 y 轴的显示范围。

#### 子图和多图布局
- subplots(): 创建一个图形和一组子图。
- subplot(): 指定当前操作的子图。
- gridspec(): 提供更灵活的子图布局。
- figure(): 创建一个新的图形。
- clf(): 清除当前图形。
- close(): 关闭一个图形窗口。

#### 样式和颜色
- colorbar(): 添加颜色条。
- colormap(): 设置颜色映射。
- style.use(): 设置绘图风格。

#### 文本和注释
- text(): 在图形上添加文本。
- annotate(): 添加注释。
- arrow(): 添加箭头。

#### 其他
- show(): 显示图形。
- savefig(): 保存图形到文件。
- suptitle(): 设置图形的超级标题。

```python
import  matplotlib.pyplot  as plt
import pandas as pd
import numpy as np
```

### 简单图例

```python
# 绘制折线图
x = [1, 2, 3, 4, 5]
x1 = [1, 3, 2, 4, 5]
y = [5, 7, 2, 8, 4]

plt.rcParams['font.family'] = 'SimHei'
plt.plot(x, y, marker='o')
plt.plot(x1, y, marker='o')
plt.xlabel('X轴标签')
plt.ylabel('Y轴标签')
plt.title('折线图示例')
# plt.show()
```

```python
x1 = [1, 2, 3, 4, 5]
x2 = [2, 4, 5, 6, 7]
y = [5, 7, 2, 8, 4]

plt.scatter(x1, y, color='red', marker='o')
plt.scatter(x2, y, color='blue', marker='o')
plt.xlabel('X轴标签')
plt.ylabel('Y轴标签')
plt.title('散点图示例')
# plt.show()
```

```python
#  条形图
categories = ['A', 'B', 'C', 'D']
values1 = [20, 45, 30, 15]
values2 = [15, 23, 17, 10]

fig, ax = plt.subplots()
bars1 = ax.bar(categories, values1)
bars2 = ax.bar(categories, values2)

# 在每个条形图上添加文本
for bar in bars1:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), 
            ha='center', va='bottom')

# 在每个条形图上添加文本
for bar in bars2:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), 
            ha='center', va='bottom')

# plt.bar(categories, values1, color='blue')
# plt.bar(categories, values2, color='red')
plt.xlabel('类别')
plt.ylabel('值')
plt.title('条形图')
# plt.show()
```

```python
#  饼图
sizes = [30, 40, 20, 10]
labels = ['A', 'B', 'C', 'D']
pie_pd=pd.DataFrame()

plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['red', 'green', 'blue', 'yellow'])
plt.title('饼图')
plt.show()
```

```python
data1 = [5, 7, 2, 8, 4, 5, 7, 8, 6, 4, 5]
data2 = [3, 5, 2, 6, 3, 1, 4, 3, 5, 1, 5]
plt.hist(data1, bins=10, color='purple', alpha=0.7)
# plt.hist(data2, bins=10, color='yellow', alpha=0.5)

plt.xlabel('值')
plt.ylabel('频率')
plt.title('直方图示例')
plt.show()
```

### 1. Matlplotlib绘图基础
> Matplotlib分为四层
- canvas（画板）：位于最底层，导入Matplotlib库时就自动存在。
- figure（画布）：建立在canvas之上，从这一层就能开始设置其参数。
- axes（子图）：将figure分成不同块，实现分面绘图。
- 图表信息（构图元素）：添加或修改axes上的图形信息，优化图表的显示效果。

```python
fig, ax = plt.subplots()
def  graph(num1,num2,num3):
    ax.barh(1, num1, color='#adb0ff')  # 绘制水平方向的条形图
    ax.barh(3, num2, color='#ffb3ff')  # 绘制水平方向的条形图
    ax.barh(5, num3, color='#90d595')  # 绘制水平方向的条形图

    ax.text(num1, 1, "广东省", size=14, weight=600, ha='right', va='bottom')
    ax.text(num2, 3, "北京市", size=14, weight=600, ha='right', va='bottom')
    ax.text(num3, 5, "云南省", size=14, weight=600, ha='right', va='bottom')
    plt.show()

graph(10,17,18)
```
